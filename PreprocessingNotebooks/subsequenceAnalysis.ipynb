```python
wordIndex = wordIndex.tolist()
```

DOING OVERLAP BASED ON POS TAGS

```python
primaryDict = defaultdict((int)) # tuple for ints of any length

for tag in all_tags_universal:
    for val in range(1,len(tag)+1):
        primaryDict[tag[0:val]] += 1
```

```python
new = []
ofLengthAtLeast = 3

for key in primaryDict.keys():
    if len(key) > ofLengthAtLeast: # long enough to prompt duplication over multiple values
        new.append((key, (primaryDict[key])))
```

```python
newMax = sorted(new, key = lambda x: x[1], reverse=True)
```
