```python
import numpy as np
import string
from collections import defaultdict
import pandas as pd
```

```python
index2word = np.load('index2word.npy')
```

```python
maxLen = 20
vocabSize = len(index2word)
print (vocabSize)
```

```python
df = pd.read_csv("questionList-index.csv")
```

```python
alpha = df.values.tolist()
```

```python
all_ques = [(x[0])[1:-1].split(', ') for x in alpha]
```

```python
allQNum = []
```

```python
for q in all_ques:
    temp = []
    
    for w in q:
        temp.append(int(w))
    
    allQNum.append(temp)
    
```

```python
countArr = np.zeros([vocabSize])
```

```python
for ques in allQNum: 
    for value in ques:
        countArr[value] = 1.0 + countArr[value]        
```

```python
rowTotal = sum(countArr)
```
